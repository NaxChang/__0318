- 建立`正弦波`的圖表範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 使用 sin() 函數計算 y 軸的值
y = np.sin(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('正弦波')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

- 同時顯示sin,cos的正弦波與餘弦波

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 使用 sin() 和 cos() 函數計算 y 軸的值
y_sin = np.sin(x)
y_cos = np.cos(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

# 設定圖表標題和軸標籤
plt.title('正弦波與餘弦波')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```
